```python
import pandas as pd
import numpy
```

```python
from time import gmtime, strftime
```

```python
def getLeagueTable():
    leagueTable = pd.read_html(io="https://www.bbc.co.uk/sport/football/premier-league/table")
    leagueTable = leagueTable[0]
    leagueTable.rename(columns = {'Unnamed: 0':'Position'}, inplace = True)
    leagueTable2 = leagueTable[['Team', 'Position', 'Pts']]
    leagueTable3 = leagueTable2[leagueTable2['Team'].notnull()]
    leagueTable3 = leagueTable3.astype( {"Position": int, "Pts": int} )
    leagueTable3['PositionDate'] = strftime("%Y-%m-%d", gmtime())
    return leagueTable3
```

```python
leagueTable = getLeagueTable()
```
